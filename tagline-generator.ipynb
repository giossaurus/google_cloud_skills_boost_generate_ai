```python
from google import genai
from google.genai import types
import base64

def generate():
  client = genai.Client(
      vertexai=True,
      project="qwiklabs-gcp-04-47dced368fb5",
      location="us-west1",
  )

  text1 = types.Part.from_text(text="""Create customizable taglines for Cymbal Direct’s new product line in partnership with an outdoor gear retailer. The taglines should include the word nature

input: Write a tagline for a durable backpack designed for hikers that makes them feel prepared. Consider styles like minimalist.
output: Built for the Journey: Your Adventure Essentials.

input: Write a tagline for a lightweight, versatile tent aimed at families that conveys a sense of connection and comfort. Consider styles like friendly.
output: Home Under the Stars: Comfort in Every Shelter.


input: Create customizable taglines for Cymbal Direct’s new product line in partnership with an outdoor gear retailer. The taglines should be influenced by the following parameters:
		•	Product Attributes: e.g., durable, lightweight
	•	Target Audience: e.g., young adventurers, families
	•	Emotional Resonance: e.g., empowered, connected
output:""")
  si_text1 = """Cymbal Direct is partnering with an outdoor gear retailer. They're launching a new line of products designed to encourage young people to explore the outdoors. Help them create catchy taglines for this product line."""

  model = "gemini-2.0-flash-001"
  contents = [
    types.Content(
      role="user",
      parts=[
        text1
      ]
    )
  ]
  generate_content_config = types.GenerateContentConfig(
    temperature = 1,
    top_p = 0.95,
    max_output_tokens = 8192,
    response_modalities = ["TEXT"],
    system_instruction=[types.Part.from_text(text=si_text1)],
  )

  for chunk in client.models.generate_content_stream(
    model = model,
    contents = contents,
    config = generate_content_config,
    ):
    print(chunk.text, end="")

generate()
```
